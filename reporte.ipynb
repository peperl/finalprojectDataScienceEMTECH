### Opción 1) Rutas de importación y exportación. Synergy logistics está considerando la posibilidad de enfocar sus esfuerzos en las 10 rutas más demandadas. Acorde a los flujos de importación y exportación, ¿cuáles son esas 10 rutas? ¿le conviene implementar esa estrategia? ¿porqué? 

#### Exportaciones

```python
valores_transporte = exports.groupby(["origin", "destination", "transport_mode"]).count()
rutas = valores_transporte['total_value'].sort_values(ascending=False)
total = rutas.sum()
rutas = rutas.head(10).reset_index()

rutas["porcentaje"] = (rutas['total_value']/total) * 100
rutas["porcentaje acum"] = rutas.cumsum()['porcentaje']

rutas
```

#### Importaciones

```python
valores_transporte = imports.groupby(["origin", "destination", "transport_mode"]).count()
rutas = valores_transporte['total_value'].sort_values(ascending=False)
total = rutas.sum()
rutas = rutas.head(10).reset_index()

rutas["porcentaje"] = (rutas['total_value']/total) * 100
rutas["porcentaje acum"] = rutas.cumsum()['porcentaje']

rutas
```

#### De acuerdo con la información es una mala estrategia ya que no se llegaría ni la mitad del mercado actual.

### Medio de transporte utilizado. ¿Cuáles son los 3 medios de transporte  más importantes para Synergy logistics considerando el valor de las importaciones y exportaciones? ¿Cuál es medio de transporte que  podrían reducir?

### Opción 3) Valor total de importaciones y exportaciones. Si Synergy Logistics quisiera enfocarse en los países que le generan el 80% del valor de las exportaciones e importaciones  ¿en qué grupo de países debería enfocar sus esfuerzos?

#### Importaciones

```python
valores_transporte = imports.groupby("origin").count()
rutas = valores_transporte.sort_values(ascending=False, by='total_value')['total_value']
total = rutas.sum()
rutas = rutas.reset_index()

rutas["porcentaje"] = (rutas['total_value']/total) * 100
rutas["porcentaje acum"] = rutas.cumsum()['porcentaje']

result = rutas[ rutas["porcentaje acum"] < 80]

result
```

##### Exportaciones

```python
valores_transporte = exports.groupby("origin").count()
rutas = valores_transporte.sort_values(ascending=False, by='total_value')['total_value']
total = rutas.sum()
rutas = rutas.reset_index()

rutas["porcentaje"] = (rutas['total_value']/total) * 100
rutas["porcentaje acum"] = rutas.cumsum()['porcentaje']

result = rutas[ rutas["porcentaje acum"] < 80]

result
```

```python
Con solamente 17 rutas generamos poco menos del 80% de la operación por lo que se considera una muy buena opción.
```
